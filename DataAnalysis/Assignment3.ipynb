```python
import pandas as pd
import numpy as np
import re
```

```python
df0 = pd.read_excel("Energy Indicators.xls") #use r before absolute file path 
df = df0[17:244]
#print(df.head())
del df['Unnamed: 0']
del df['Unnamed: 1']
df.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
#print(df.head(50))
df.replace('...', np.nan,  inplace=True)
#df['Country'] = df['Country'].apply(lambda x: re.findall('([A-Za-z, \- ]*)',x)[0].strip())
#print(df.head(50))

df['Country'] = df['Country'].apply(lambda x: re.findall('([^0-9]*)',x)[0].strip())
df['Country'] = df['Country'].apply(lambda x: re.findall('([^\(]*)',x)[0].strip())

dict1 ={"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong",
"China, Macao Special Administrative Region": "Macao"}
 

#print(df.head(100))
df['Country'].replace(dict1, None,  inplace=True)

Energy = df
print(len(Energy))
print(len(Energy['Country'].unique()))
Energy=Energy.set_index('Country')
Energy['Energy Supply']=1000000*Energy['Energy Supply']
df['Country'].unique()
```

```python
Energy.groupby(Energy.index).size().reset_index(name='counts').sort_values('counts')
```

```python
df = pd.read_csv('world_bank.csv', header=4)
df.replace('...', np.nan,  inplace=True)

df['Country Name'] = df['Country Name'].apply(lambda x: re.findall('([^0-9]*)',x)[0].strip())

dict1 ={"Korea, Rep.": "South Korea", 
"Iran, Islamic Rep.": "Iran",
"Hong Kong SAR, China": "Hong Kong"}
df['Country Name'].replace(dict1, None,  inplace=True)

dict1 ={"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong",
"China, Macao Special Administrative Region": "Macao"}
df['Country Name'].replace(dict1, None,  inplace=True)

#df['Country Name'] = df['Country Name'].apply(lambda x: re.findall('([A-Za-z \- \. \& \( \) \, ]*)',x)[0].strip())

#df['Country Name'] = df['Country Name'].apply(lambda x: re.findall('([^\(]*)',x)[0].strip())
GDP = df
print(len(GDP))
print(len(GDP['Country Name'].unique()))
GDP=GDP.set_index('Country Name')
GDP
GDP.groupby(GDP.index).size().reset_index(name='counts').sort_values('counts')
df['Country Name'].unique()
```
